## Notes:
### An example that helped me remember copula is from [1]. 

1. The quoted example is:

"Let’s start with an example problem case. Say we measure two variables that are non-normally distributed and correlated. For example, we look at various rivers and for every river we look at the maximum level of that river over a certain time-period. In addition, we also count how many months each river caused flooding. For the probability distribution of the maximum level of the river we can look to Extreme Value Theory which tells us that maximums are Gumbel distributed. How many times flooding occured will be modeled according to a Beta distribution which just tells us the probability of flooding to occur as a function of how many times flooding vs non-flooding occured.

It’s pretty reasonable to assume that the maximum level and number of floodings is going to be correlated. However, here we run into a problem: how should we model that probability distribution? Above we only specified the distributions for the individual variables, irrespective of the other one (i.e. the marginals). In reality we are dealing with a joint distribution of both of these together.' [1]

2. Transforming uniform rv to normal and back
Take uniform RV and find the inv CDF of that value. Remember that CDF runs from 0 to 1 and so does our standard uniform.

3. So there is a reason why Gaussian Copulae don't really work- it assumes that the correlation we give is constant and symmetric. [1] and [2] both address why this led to a lot of issues. Gaussian copulae do not assume that one default is independent of another default. But it assumes that that one default occuring does not change the correlation value. It does. One default happening is going to change the other defaults happening.



```python
import seaborn as sns
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Generating Uniform RV
x = stats.uniform(0,1).rvs(10000)
sns.displot(x, kind="hist")
```

```python
# Inverse CDF of normal of x
norm = stats.distributions.norm()
x_inv = norm.ppf(x)
sns.displot(x_inv, kind = 'hist')
```

```python
# plotting both distributions together to find how it varies and what the transformation does
data = pd.DataFrame({'x': x, 'x_inv': x_inv})

# Create the joint plot
h = sns.jointplot(data=data, x='x', y='x_inv')
```

```python
# And for beta:
beta = stats.distributions.beta(a=10,b=3)
x_inv1 = beta.ppf(x)

data1 = pd.DataFrame({'x':x, 'x_inv1':x_inv1})

h1 = sns.jointplot(data=data1,x='x',y='x_inv1')
```

```python
# And for Gumbel:
gumbel = stats.distributions.gumbel_l()
x_trans = gumbel.ppf(x)

data2 = pd.DataFrame({'x':x, 'x_trans':x_trans})

h2 = sns.jointplot(data=data2,x='x',y='x_trans')
```

In order to do the opposite transformation from an arbitrary distribution to the uniform(0, 1) we just apply the inverse of the inverse CDF – the CDF:

```python
x_inv_inv = norm.cdf(x_inv)

data = pd.DataFrame({'x_inv': x_inv, 'x_inv_inv': x_inv_inv})

# Create the joint plot
h = sns.jointplot(data=data, x='x_inv', y='x_inv_inv')
```

## So finally: Copulae
Multivariate normal distributions (correlated) -> Taking CDF to find uniform distributions -> transform to whatever function you need!

```python
mvnorm = stats.multivariate_normal(mean=[0, 0], cov=[[1., 0.5], 
                                                     [0.5, 1.]])
# Generate random samples from multivariate normal with correlation .5
x = mvnorm.rvs(100000)
```

```python
x[1]
```

```python
data_mv = pd.DataFrame({'x1':x[:,0],'x2':x[:,1]})
h = sns.jointplot(data=data_mv,x='x1',y='x2')
```

```python
sns.displot(data=data_mv,x='x1',y='x2')
```

```python
norm = stats.norm()
x_unif = norm.cdf(x)
```

```python
x_unif
```

```python
data = pd.DataFrame({'x1_unif': x_unif[:,0], 'x2_unif': x_unif[:,1]})

# Create the joint plot
h = sns.jointplot(data=data, x='x1_unif', y='x2_unif', kind = "hex")
```

```python
# Transform to Beta and Gumbel
m1 = stats.gumbel_l()
m2 = stats.beta(a=10, b=2)

x1_trans = m1.ppf(x_unif[:, 0])
x2_trans = m2.ppf(x_unif[:, 1])
```

```python
data_final = pd.DataFrame({'x1_trans':x1_trans,'x2_trans':x2_trans})
```

```python
sns.jointplot(data=data_final, x='x1_trans', y='x2_trans', kind = "hist")
```

```python
sns.jointplot(data=data_final, x='x1_trans', y='x2_trans',kind='kde',fill=True,xlim=(-6, 2), ylim=(.6, 1.0))
```
